```python
import cv2
import matplotlib.pyplot as plt
def Eye_Detection(img):
    img = cv2.imread(img)
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    eye = eye_cascade.detectMultiScale(img, 1.3, 20)
    for (x, y, w, h) in eye:
        cv2.rectangle(img, (x,y), (x+w, y+h),(0,255,0),2)
    plt.matshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
im_lis=['Woman-with-soft-round-face-wearing-square-eyeglasses-frame.jpg', 
       'Beautiful-smile.jpg',
       'Jonah-Hill.webp',
        'VAI_01-22_People_Panther_JPG_white_hor_007-800x460.webp',
       'vag.webp']
for i in im_lis:
    Eye_Detection(i)
```
